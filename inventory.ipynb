# 🧮 Step 5: Add New Column (Total Value)
## 👉 Total Value = Stock × Price

```python
df["TotalValue"] = df["Stock"] * df["Price"]
df
```

# ⚠️ Step 6: Find Low Stock Items

## 👉 Example: Stock < 50

```python
low_stock = df[df["Stock"] < 50]
low_stock
```

# Step 7: Category-wise Analysis

```python
category_wise = df.groupby("Category")["TotalValue"].sum()
category_wise
```

# 🏆 Step 8: Top Expensive Products

```python
top_products = df.sort_values(by="Price",ascending = False)
top_products
```

```python
df.loc[df["ProductName"] == "Laptop", "Stock"] = 45
df
```

```python
df.loc[df["Category"] == "Furniture", "Stock"] = 150
df
```

```python
df.loc[df["ProductName"] == "Desk", "Stock"] = 70
df
```
